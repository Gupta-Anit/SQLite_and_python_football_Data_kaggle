##### Question 1 (20 points): Write a SQL query that lists all the players born between 1987 and 1990 inclusive, sort them from the oldest to the youngest. The output of this query should be of the form:

##### Player Name | Birthday


```python
import os
import sqlite3
import pandas as pd
con = sqlite3.connect('database.sqlite')
cursorObj = con.cursor()
```

```python
df0 = pd.read_sql_query("SELECT Player_Name as 'Player Name', Birthday \
                            from Player \
                            where birthday between '1987-01-01' and '1990-12-31' \
                            order by birthday ;",con)
df0
```

##### Country |  League Name | Total Goals Scored

#### Question 5 (40 points): Write a SINGLE SQL query that finds the date that had the most goals scored on, per each different season and league. The output of this query should be of the form:

##### Date (dd/mm/yy) | Season | League Name | Goals scored

```python

df4 = pd.read_sql_query("select substr(date,0,7)||substr(date,9) as 'Date (dd/mm/yy)', Season as Season, league_name as 'League Name', maX( total_goal_scored) as 'Goals Scored' from \
                                (select strftime('%d/%m/%Y',date) as date,c.name as country,season,l.name as league_name, sum(home_team_goal+away_team_goal) as total_goal_scored\
                                        from Match m, country c, league l\
                                        where c.id=l.country_id and \
                                              m.league_id=l.id  \
                                        group by country,season,date\
                                        order by country ,total_goal_Scored desc)\
                        group by country,season\
                        order by country \
                        ", con)
df4
```
